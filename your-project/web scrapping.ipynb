```python
import requests 

from bs4 import BeautifulSoup 
```

```python
#Download the webpage:

url = "https://www.studying-in-uk.org/list-of-universities-in-the-united-kingdom/" 

page = requests.get(url)


#Parse the html:

ukunis = BeautifulSoup(page.content, 'html.parser')
```

```python
# print the formatted html

print(ukunis.prettify())
```

```python
ukunis.select("h3")[0]
```

```python
#we need to get the recognised bodies

ukunis.select("ul")[3].select("li")
```

```python
# number of universities listed as recognised bodies

uni_length = len(ukunis.select("ul")[3].select("li"))
```

```python
# how to access each individual university name in the list

ukunis.select("ul")[3].select("li")[0].get_text()
```

```python
# Let's generate a loop to obtain scrappe the university names
    
uk_uni_rb = [ukunis.select("ul")[3].select("li")[i].get_text() for i in range (uni_length)]
```
